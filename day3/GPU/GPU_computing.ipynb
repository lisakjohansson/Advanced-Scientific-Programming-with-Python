```python
import numpy as np
import cupy as cp
```

float64

```python
sizes = [128, 256, 512, 1024, 2048]

def numpy_fft(a):
    return np.fft.fft2(a)

def cupy_fft(a):
    out = cp.fft.fft2(a)
    cp.cuda.Stream.null.synchronize()
    return out

for n in sizes:
    print(f"\nStorlek: {n}x{n}")

    a_np = np.random.random((n, n))
    a_cp = cp.asarray(a_np)

    # warm up for GPU
    cupy_fft(a_cp)

    print("NumPy:")
    %timeit -n 10 -r 5 numpy_fft(a_np)

    print("CuPy:")
    %timeit -n 10 -r 5 cupy_fft(a_cp)
```

float32

```python
sizes = [128, 256, 512, 1024, 2048]

for n in sizes:
    print(f"\nStorlek: {n}x{n} med float32")

    a_np = np.random.random((n, n)).astype(np.float32)
    a_cp = cp.asarray(a_np)

    # warm up for GPU
    cupy_fft(a_cp)

    print("NumPy:")
    %timeit -n 10 -r 5 numpy_fft(a_np)

    print("CuPy:")
    %timeit -n 10 -r 5 cupy_fft(a_cp)
```
