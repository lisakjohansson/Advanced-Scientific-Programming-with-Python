```python
import numpy as np
```

```python
# a. Create a null vector of size 10 but the fifth value which is 1

N = 10
ind = 5 - 1 # zero based
null_vector = np.zeros(N)
null_vector[ind] = 1
print(null_vector)
```

```python
# b. Create a vector with values ranging from 10 to 49

vector = np.array(range(10, 50))
print(vector)
```

```python
# c. Reverse a vector (first element becomes last)

reversed_vector = np.array(list(reversed(vector)))
print(reversed_vector)
```

```python
# d. Create a 3x3 matrix with values ranging from 0 to 8

matrix = np.arange(9).reshape(3, 3)
print(matrix)
```

```python
# e. Find indices of non-zero elements from [1,2,0,0,4,0]

vector = np.array([1,2,0,0,4,0])
non_zero_elements_indices = np.nonzero(vector)
print(non_zero_elements_indices)
```

```python
# f. Create a random vector of size 30 and find the mean value

random_vector = np.random.random(30)
mean_value = random_vector.mean()
print(mean_value)
```

```python
# g. Create a 2d array with 1 on the border and 0 inside

arr = np.ones((5, 5))

# Slicing [rows, columns]
# Start at row 1, stop before the last row
# Start at column 1, stop before last column
arr[1:-1, 1:-1] = 0

print(arr)
```

```python
# h. Create a 8x8 matrix and fill it with a checkerboard pattern

checkerboard_matrix = np.zeros(64).reshape(8, 8)
checkerboard_matrix[1::2, ::2] = 1
checkerboard_matrix[::2, 1::2] = 1

print(checkerboard_matrix)
```

```python
# i. Create a checkerboard 8x8 matrix using the tile function

checkerboard_matrix = np.tile([[0, 1], [1, 0]], (4, 4))
print(checkerboard_matrix)
```

```python
# j. Given a 1D array, negate all elements which are between 3 and 8, in place

Z = np.arange(11)
print(Z)

Z[(Z > 3) & (Z < 8)] *= -1 # Create boolean mask
print(Z)
```

```python
# k. Create a random vector of size 10 and sort it

Z = np.random.random(10)
print(Z)

Z = np.sort(Z)
print("Sorted Z from smallest to largest: \n", Z)
```

```python
# l. Consider two random array A anb B, check if they are equal

A = np.random.randint(0,2,5)
print(A)
B = np.random.randint(0,2,5)
print(B)

equal = np.array_equal(A, B)
print(equal)
```

```python
# m. How to calculate the square of every number in an array in place (without creating temporaries)?

Z = np.arange(10, dtype=np.int32)
print(Z.dtype)
print(Z)

np.square(Z, out = Z)
print(Z)
print(Z.dtype)
```

```python
# n. How to get the diagonal of a dot product?

A = np.arange(9).reshape(3,3)
B = A + 1

C = np.dot(A,B)

D = np.diagonal(C)
print(D)
```
